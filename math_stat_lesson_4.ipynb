Задание 1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
Найдите ее среднее значение и дисперсию.

Распределение вероятностей случайной величины $X$ в зависимости от плотности:
$$f(x) = \begin{cases} 0, если \; x \le a; \\ \frac{1}{b - a}, если \; a < x \le b; \\ 0, \; если \; x > b. \end{cases}$$

Среднее $M(X)$: 
$$M(X) = \frac{a + b}{2}=\frac{200+800}{2}=\frac{1000}{2}=500$$

Дисперсия $D(X)$:
$$D(X) = \frac{(b - a)^2}{12}=\frac{(800 - 200)^2}{12}=\frac{(600)^2}{12}=\frac{360000}{12}=30000$$

Задание 5. На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

Дисперсия $ \sigma^2 = D(X)$, т.е среднеквадратичное отклонение $\sigma= \sqrt{D(X)}=\sqrt{25}=5$
<br>Посчитаем отклонение в $\sigma$: <br>$\frac{190-M(X)}{\sigma}=\frac{190-178}{5}=\frac{12}{5}=2.4$

Итого рост человека равный $190$см отклоняется на $2.4 \sigma$
